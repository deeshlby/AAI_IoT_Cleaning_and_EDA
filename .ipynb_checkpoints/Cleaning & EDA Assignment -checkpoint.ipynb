```python
#make a copy of the raw data so that we can go back and refer to it later
df = df_raw.copy()
```

```python

# this line of code is from module 2 lab session
#df['Date'] = df['Date'].apply(pd.to_datetime)
#df.dtypes
df['Datetime'] = df[['Date', 'Time']].apply(lambda x: " ".join(x.values.astype(str)), axis="columns")
```

```python
df.head()
```

```python
#create your Datetime column
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

```python
df.dtypes
```

```python
df["Datetime"][0]
```

```python
#convert all data columns to numeric types
# https://www.statology.org/pandas-convert-object-to-float/
df['Global_active_power'] = df['Global_active_power'].replace('?', '').apply(pd.to_numeric)
df['Global_reactive_power'] = df['Global_reactive_power'].replace('?', '').apply(pd.to_numeric)
df['Voltage'] = df['Voltage'].replace('?', '').apply(pd.to_numeric)
df['Global_intensity'] = df['Global_intensity'].replace('?', '').apply(pd.to_numeric)
df['Sub_metering_1'] = df['Sub_metering_1'].replace('?', '').apply(pd.to_numeric)
df['Sub_metering_2'] = df['Sub_metering_2'].replace('?', '').apply(pd.to_numeric)
df.dtypes
```

Let's use the Datetime column to turn the Date and Time columns into date and time dtypes.

```python
df['Date'] = df['Datetime'].dt.date
df['Time'] = df['Datetime'].dt.time
```

```python
df.Date[0]
```

Now that we've got the data in the right datatypes, let's take a look at the describe() results

```python
#use datetime_is_numeric = True to get statistics on the datetime column
desc = df.describe(datetime_is_numeric = True)

#force the printout not to use scientific notation
desc[desc.columns[:-1]] = desc[desc.columns[:-1]].apply(lambda x: x.apply("{0:.4f}".format))
desc
```

## Data Covariance and Correlation

Let's take a look at the Correlation Matrix for the four global power variables in the dataset.

```python
axes = pd.plotting.scatter_matrix(df[['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity']], alpha=0.5,figsize = [10,10])
corr = df[['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity']].corr(method = 'spearman').to_numpy() #nonlinear
for i, j in zip(*plt.np.triu_indices_from(axes, k=1)):
    axes[i, j].annotate("%.3f" %corr[i,j], (0.8, 0.8), xycoords='axes fraction', ha='center', va='center')
plt.show()
```

**Q: Describe any patterns and correlations that you see in the data. What effect does this have on how we use this data in downstream tasks?**

A: Looking at the correlation covariance scatterplots and the Spearman rho correlation matrices we can see that the Global_active_power and the Global_intensity are highly correlated with a matrix of 0.995 almost a 1. While we can not see any other meaningful correlation between any other global power variables. 
The high correlation between the Global intensity and the Global active power means that they can be interchangable when it comes to predicting the Sub metering values. So the Global intensity, Global reactive power, and Voltage can be the independant variables used our ML models to predict the Sub metering values for the household.
